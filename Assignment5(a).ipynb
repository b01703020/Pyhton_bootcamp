Q7

Extract two PCA factors from the payment data subset containing the following 18 columns only: â€˜PAY_0â€™ to 'PAY_AMT6. Concatenate these two PCA factors with the following columns from the original dataframe: â€˜LIMIT_BALâ€™ and â€˜AGEâ€™ columns.

```python
X_subset = X.drop(['LIMIT_BAL','AGE'],axis = 1).copy()
X_subset
```

```python
from sklearn.decomposition import PCA
model = PCA(n_components=2)
model.fit(X_subset)
X_subset_pca = model.transform(X_subset)
```

```python
X_new = X[['LIMIT_BAL','AGE']].copy()
X_new['pca1'] = X_subset_pca[:,0].copy()
X_new['pca2'] = X_subset_pca[:,1].copy()
X_new
```

Q8

Re-run the Gaussian Naive Bayes algorithm to classify labels into default(1)or notd efault (0) but where now the features contain â€˜LIMIT_BALâ€™, â€˜AGEâ€™ and the two payment PCA factors.

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X_new, y, random_state=0)

model_new = GaussianNB()
model_new.fit(Xtrain, ytrain)
y_model = model_new.predict(Xtest)
```
