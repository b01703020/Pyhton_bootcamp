Q6

Set up an array x = [1,2,...,1000] and show that the following formula holds for N = 1000:

$ \sum\limits_{i = 1} ^{N} i = N*(N+1)/2$

```python
def sum_integer_from_1_to_(N):
    return N*(N+1)/2

array_from_1_to_1000 = np.arange(1,1001)

np.sum(array_from_1_to_1000) == sum_integer_from_1_to_(1000)
```

Q7

Compute the cumulative sum in Problem 6 using (i) a for loop and (ii) NumPy’s cumsum function. 

Time both using %timeit and report your findings.

```python
def csum_using_for(arr):
    '''cumulative sum of an array using for loop'''
    temp = 0
    for i in range(len(arr)):    
        arr[i] += temp
        temp = arr[i]        
    return arr

array_from_1_to_1000 = np.arange(1,1001)
csum_using_for(array_from_1_to_1000)
```

```python
array_from_1_to_1000 = np.arange(1,1001)
%timeit csum_using_for(array_from_1_to_1000)
```

```python
array_from_1_to_1000 = np.arange(1,1001)
np.cumsum(array_from_1_to_1000)
```

```python
array_from_1_to_1000 = np.arange(1,1001)
%timeit np.cumsum(array_from_1_to_1000)
```
